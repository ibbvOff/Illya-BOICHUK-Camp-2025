```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, Lasso, LinearRegression
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score

diabetes = load_diabetes()
X, y = diabetes.data, diabetes.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
scaler = StandardScaler()
X_train, X_test = scaler.fit_transform(X_train), scaler.transform(X_test)

def plot_results(y_test, y_pred, model_name):
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.scatter(y_test, y_pred, color='blue', alpha=0.7)
    plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')
    plt.title(f'{model_name}: Actual vs Predicted')
    plt.xlabel('Actual Values')
    plt.ylabel('Predicted Values')
    plt.subplot(1, 2, 2)
    plt.scatter(y_pred, y_pred - y_test, color='blue', alpha=0.7)
    plt.axhline(y=0, color='red', linestyle='--')
    plt.title(f'{model_name}: Residuals')
    plt.xlabel('Predicted Values')
    plt.ylabel('Residuals')
    plt.tight_layout()
    plt.show()
```

### Ridge

```python
ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)
y_pred_ridge = ridge.predict(X_test)
ridge_r2 = r2_score(y_test, y_pred_ridge)
ridge_mse = mean_squared_error(y_test, y_pred_ridge)
plot_results(y_test, y_pred_ridge, "Ridge Regression")
print(f"Ridge Regression: R2 = {ridge_r2:.4f}, MSE = {ridge_mse:.2f}")
```

### Lasso

```python
lasso = Lasso(alpha=0.01, max_iter=10000)
lasso.fit(X_train, y_train)
y_pred_lasso = lasso.predict(X_test)
lasso_r2 = r2_score(y_test, y_pred_lasso)
lasso_mse = mean_squared_error(y_test, y_pred_lasso)
plot_results(y_test, y_pred_lasso, "Lasso Regression")
print(f"Lasso Regression: R2 = {lasso_r2:.4f}, MSE = {lasso_mse:.2f}")
```

### Polynomial

```python
degree = 2
poly = PolynomialFeatures(degree, include_bias=False)
X_train_poly, X_test_poly = poly.fit_transform(X_train), poly.transform(X_test)
poly_model = LinearRegression()
poly_model.fit(X_train_poly, y_train)
y_pred_poly = poly_model.predict(X_test_poly)
poly_r2 = r2_score(y_test, y_pred_poly)
poly_mse = mean_squared_error(y_test, y_pred_poly)
plot_results(y_test, y_pred_poly, f"Polynomial Regression (degree={degree})")
print(f"Polynomial Regression (degree={degree}): R2 = {poly_r2:.4f}, MSE = {poly_mse:.2f}")
```

### Polynomial + Linear

```python
poly_linear = LinearRegression()
poly_linear.fit(X_train_poly, y_train)
y_pred_poly_linear = poly_linear.predict(X_test_poly)
poly_linear_r2 = r2_score(y_test, y_pred_poly_linear)
poly_linear_mse = mean_squared_error(y_test, y_pred_poly_linear)
plot_results(y_test, y_pred_poly_linear, "Polynomial + Linear Regression")
print(f"Polynomial + Linear Regression: R2 = {poly_linear_r2:.4f}, MSE = {poly_linear_mse:.2f}")
```

### Polynomial + Ridge

```python
poly_ridge = Ridge(alpha=1.0)
poly_ridge.fit(X_train_poly, y_train)
y_pred_poly_ridge = poly_ridge.predict(X_test_poly)
poly_ridge_r2 = r2_score(y_test, y_pred_poly_ridge)
poly_ridge_mse = mean_squared_error(y_test, y_pred_poly_ridge)
plot_results(y_test, y_pred_poly_ridge, "Polynomial + Ridge Regression")
print(f"Polynomial + Ridge Regression: R2 = {poly_ridge_r2:.4f}, MSE = {poly_ridge_mse:.2f}")
```

### Polynomial + Lasso

```python
poly_lasso = Lasso(alpha=0.1, max_iter=10000)
poly_lasso.fit(X_train_poly, y_train)
y_pred_poly_lasso = poly_lasso.predict(X_test_poly)
poly_lasso_r2 = r2_score(y_test, y_pred_poly_lasso)
poly_lasso_mse = mean_squared_error(y_test, y_pred_poly_lasso)
plot_results(y_test, y_pred_poly_lasso, "Polynomial + Lasso Regression")
print(f"Polynomial + Lasso Regression: R2 = {poly_lasso_r2:.4f}, MSE = {poly_lasso_mse:.2f}")
```

### Ridge + Linear

```python
ridge_linear = Ridge(alpha=1.0)
ridge_linear.fit(X_train, y_train)
y_pred_ridge_linear = ridge_linear.predict(X_test)
ridge_linear_r2 = r2_score(y_test, y_pred_ridge_linear)
ridge_linear_mse = mean_squared_error(y_test, y_pred_ridge_linear)
plot_results(y_test, y_pred_ridge_linear, "Ridge + Linear Regression")
print(f"Ridge + Linear Regression: R2 = {ridge_linear_r2:.4f}, MSE = {ridge_linear_mse:.2f}")
```
