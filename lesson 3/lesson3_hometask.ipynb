## Answer 1

```python
import pandas as pd
import numpy as np

def answer_one():
    energy_data = pd.read_excel('data/Energy Indicators.xls', usecols="C:F")
    energy_df = pd.DataFrame(energy_data)
    energy_df.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
    energy_df = energy_df[17:244]
    energy_df['Energy Supply'] = energy_df['Energy Supply'].apply(lambda x: x * 1_000_000 if isinstance(x, int) else x)
    energy_df = energy_df.replace('...', np.nan).infer_objects(copy=False)
    
    country_name_mapping = {
        'Republic of Korea': 'South Korea',
        'United States of America': 'United States',
        'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
        'China, Hong Kong Special Administrative Region': 'Hong Kong'
    }
    
    energy_df['Country'] = energy_df['Country'].str.replace(r' \(.*\)|[0-9]', '', regex=True)
    energy_df['Country'] = energy_df['Country'].replace(country_name_mapping)

    gdp_data = pd.read_excel('data/world_bank.xls', sheet_name='Data', skiprows=3)
    gdp_df = pd.DataFrame(gdp_data)
    
    gdp_country_mapping = {
        'Korea, Rep.': 'South Korea',
        'Iran, Islamic Rep.': 'Iran',
        'Hong Kong SAR, China': 'Hong Kong'
    }
    
    gdp_df['Country Name'] = gdp_df['Country Name'].replace(gdp_country_mapping)
    gdp_df = gdp_df.drop(gdp_df.columns[1:50], axis=1)
    gdp_df = gdp_df.drop(gdp_df.columns[11:18], axis=1)

    scimago_data = pd.read_excel('data/scimagojr.xlsx')
    scimago_df = pd.DataFrame(scimago_data)
    scimago_df = scimago_df[scimago_df['Rank'] <= 15]

    merged_df = pd.merge(scimago_df, energy_df, on='Country', how='inner')
    merged_df = pd.merge(merged_df, gdp_df, left_on='Country', right_on='Country Name', how='inner')
    
    merged_df.drop('Country Name', axis=1, inplace=True)
    merged_df.drop('Region', axis=1, inplace=True)
    merged_df.set_index('Country', inplace=True)
    
    return merged_df

answer_one()
```

# Answer 2

```python
def answer_two():
    Top15 = answer_one()
    avgGDP = Top15.loc[:, '2006':'2015'].mean(axis=1).sort_values(ascending=False)
    return avgGDP
answer_two()
```

# Answer 3

```python
def answer_three():
    Top15 = answer_one()
    avgGDP = answer_two()
    sixth_largest_country = avgGDP.index[5]
    gdp_change = Top15.loc[sixth_largest_country, '2015'] - Top15.loc[sixth_largest_country, '2006']
    return gdp_change
answer_three()
```

# Answer 4

```python
def answer_four():
    Top15 = answer_one()
    Top15['Citation Ratio'] = Top15['Citations'] / Top15['Self-citations']
    max_ratio_country = Top15['Citation Ratio'].idxmax()
    max_ratio_value = Top15['Citation Ratio'].max()
    
    return (max_ratio_country, max_ratio_value)
answer_four()
```

# Answer 5

```python
def answer_five():
    Top15 = answer_one()
    Top15['Population Estimate'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    Top15 = Top15.sort_values(by='Population Estimate', ascending=False)
    third_most_populous_country = Top15.index[2]

    return third_most_populous_country
answer_five()
```

# Answer 6

```python
def answer_six():
    Top15 = answer_one()
    Top15['Citable Documents per Capita'] = Top15['Citable documents'] / (Top15['Energy Supply'] / Top15['Energy Supply per Capita'])
    correlation = Top15['Citable Documents per Capita'].corr(Top15['Energy Supply per Capita'])
    return correlation
answer_six()
```

# Answer 7

```python
def answer_seven():
    Top15 = answer_one()
    ContinentDict = {'China':'Asia', 
                     'United States':'North America', 
                     'Japan':'Asia', 
                     'United Kingdom':'Europe', 
                     'Russian Federation':'Europe', 
                     'Canada':'North America', 
                     'Germany':'Europe', 
                     'India':'Asia',
                     'France':'Europe', 
                     'South Korea':'Asia', 
                     'Italy':'Europe', 
                     'Spain':'Europe', 
                     'Iran':'Asia',
                     'Australia':'Australia', 
                     'Brazil':'South America'}
    
    continent_df = pd.DataFrame(list(ContinentDict.items()), columns=['Country', 'Continent'])
    Top15 = pd.merge(Top15, continent_df, on='Country')
    Top15['Population Estimate'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    result = Top15.groupby('Continent')['Population Estimate'].agg(['size', 'sum', 'mean', 'std'])
    return result
answer_seven()
```
