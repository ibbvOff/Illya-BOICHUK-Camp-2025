```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering, KMeans, DBSCAN
from scipy.cluster.hierarchy import dendrogram, linkage

X, y = make_blobs(n_samples = 500, n_features = 2, centers = 5, 
                           cluster_std = 0.6, random_state = 0)
colors = np.array([plt.cm.Spectral(val)
          for val in np.linspace(0, 1, len(set(y)))])
plt.figure(figsize=(8,6))
plt.title("Generated Data")
plt.scatter(X[:,0], X[:,1], c= colors[y], s= 20)    
```

```python
model_1 = AgglomerativeClustering(n_clusters=5, linkage='single')
labels_1 = model_1.fit_predict(X)
plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=labels_1, cmap='Spectral', s=20)
plt.title("Agglomerative Clustering (Single Linkage)")
plt.show()
```

```python
model_2 = AgglomerativeClustering(n_clusters=5, linkage='complete')
labels_2 = model_2.fit_predict(X)
plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=labels_2, cmap='Spectral', s=20)
plt.title("Agglomerative Clustering (Complete Linkage)")
plt.show()
```

```python
linked = linkage(X, method='complete')  
plt.figure(figsize=(12, 6))
dendrogram(linked, truncate_mode='lastp', p=30, show_leaf_counts=True)
plt.title("Dendrogram (Complete Linkage)")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.show()
```

```python
db = DBSCAN(eps=0.6, min_samples=5)
labels_db = db.fit_predict(X)
plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=labels_db, cmap='Spectral', s=20)
plt.title("DBSCAN Clustering")
plt.show()
```

```python
inertia = []
K = range(1, 10)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8,6))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=0)
labels_kmeans = kmeans.fit_predict(X)
plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=labels_kmeans, cmap='Spectral', s=20)
plt.title("So best KMeans Clustering (k=5)")
plt.show()
```
